```python
#importing libraries
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading dataset
data=pd.read_csv('hr_data.csv')
data
```

```python
#loading excel file of employee satisfaction_evalution
df2=pd.read_excel('employee_satisfaction_evaluation.xlsx')
df2
```

```python
main_df=data.set_index('employee_id').join(df2.set_index('EMPLOYEE #'))
```

```python
main_df.reset_index()
```

```python
#correalation
```

```python
corr=main_df.corr()
fig,ax=plt.subplots(figsize=(10,10))
ax.legend()
cax=ax.matshow(corr)
fig.colorbar(cax)
plt.xticks(range(len(corr.columns)),corr.columns,rotation='vertical')
plt.yticks(range(len(corr.columns)),corr.columns)
plt.show()
```

```python
sns.barplot(x='left',y='satisfaction_level',data=main_df)
plt.show()
```

```python
sns.barplot(x='number_project',y='satisfaction_level',data=main_df)
plt.show()
```

```python
sns.barplot(x='promotion_last_5years',y='satisfaction_level',data=main_df,hue='left')
plt.show()
```

```python
sns.pairplot(main_df)
plt.show()
```
